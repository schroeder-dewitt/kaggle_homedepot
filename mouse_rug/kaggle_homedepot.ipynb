```python
import pandas as pd

train = pd.read_csv("../data/train.csv")
attributes = pd.read_csv("../data/attributes.csv")
descriptions = pd.read_csv("../data/product_descriptions.csv")
all = train.merge(attributes, on="product_uid").merge(descriptions, on="product_uid")
all_grouped = all.groupby("product_uid")
```
